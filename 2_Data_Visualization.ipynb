<h2>Bagaimana Performa Penjualan dan Revenue Perusahaan dalam Beberapa Bulan Terakhir?</h2>

Seperti biasa, library tercinta kita yaitu pandas telah menyediakan sebuah method bernama resample(). Method ini memungkinkan kita untuk mengubah frekuensi atau melakukan resampling terhadap DataFrame yang memiliki komponen time series. Untuk menggunakan method ini, kita harus mendefinisikan parameter rule (mengatur target konversi) dan on (nama kolom bertipe datetime yang akan diubah frekuensinya). Selengkapnya, Anda dapat membaca dokumentasi berikut: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

all_df = pd.read_csv('all.csv')
all_df.head()
```

```python
datetime_columns = ["order_date", "delivery_date"]
 
for column in datetime_columns:
  all_df[column] = pd.to_datetime(all_df[column])

all_df.info()
```

```python
all_df.info()
```

```python
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)
monthly_orders_df.head()
```
